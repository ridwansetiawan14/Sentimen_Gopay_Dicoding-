```python
!pip install google-play-scraper
!pip install pandas
from google_play_scraper import Sort, reviews
import pandas as pd

app_id = 'com.gojek.gopay'  # ID untuk Gopay
lang = 'id'  # Bahasa Indonesia
country = 'id'  # Ulasan dari pengguna di Indonesia
batch_size = 2000  # Ukuran batch per request
total_reviews_to_scrape = 12500  # Jumlah ulasan yang ingin diambil

# List untuk menyimpan semua ulasan
all_reviews = []

# Variabel untuk melacak continuation_token
continuation_token = None

# Loop untuk mengambil ulasan dalam beberapa batch
while len(all_reviews) < total_reviews_to_scrape:
    result, continuation_token = reviews(
        app_id,
        lang=lang,
        country=country,
        sort=Sort.NEWEST,  # Mengambil ulasan terbaru, bisa juga Sort.MOST_RELEVANT atau Sort.RATING
        count=batch_size,
        continuation_token=continuation_token  # Lanjutkan scraping dari token terakhir
    )

    # Menambahkan ulasan ke dalam list
    all_reviews.extend(result)

    print(f'Scraped {len(all_reviews)} reviews so far...')

    # Jika tidak ada lagi continuation_token, berhenti
    if continuation_token is None:
        break

# Potong ulasan jika lebih dari yang diinginkan
all_reviews = all_reviews[:total_reviews_to_scrape]

# Konversi ke DataFrame
df = pd.DataFrame(all_reviews)

# Simpan ulasan ke file CSV
df.to_csv('gopayreviews.csv', index=False)

print(f'Successfully scraped {len(df)} reviews!')
# Menampilkan 5 baris pertama dari DataFrame
df.head()
```
